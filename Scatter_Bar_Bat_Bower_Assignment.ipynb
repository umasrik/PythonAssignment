# Generate Scatter Plot and Bar chart by using cleaned Bowler dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data loaded into a DataFrame named 'df'
# Read the bowlers dataset which is downloaded from github(File path)
df=pd.read_csv(r"C:\Users\navee\Downloads\Bowler_data.csv\Bowler_data_clean.csv")

print(df)
```

# Scatter Plot 

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataset is stored in a DataFrame called df
# If it's stored in a CSV file, you can read it using pd.read_csv('filename.csv')
# For this example, let's assume the DataFrame is already created and named df

# Remove rows with missing values in 'Overs' and 'Wkts' columns
df_cleaned = df.dropna(subset=['Overs', 'Wkts'])

# Convert 'Overs' and 'Wkts' columns to numeric (ignoring errors)
df_cleaned['Overs'] = pd.to_numeric(df_cleaned['Overs'], errors='coerce')
df_cleaned['Wkts'] = pd.to_numeric(df_cleaned['Wkts'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Overs'], df_cleaned['Wkts'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Overs vs Wickets')
plt.xlabel('Overs')
plt.ylabel('Wickets')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Bar chart 

```python
import matplotlib.pyplot as plt
import pandas as pd

# Remove rows with missing values in 'Wkts' and 'Bowler' columns
df_cleaned = df.dropna(subset=['Wkts', 'Bowler'])

# Group by 'Bowler' and sum the total wickets taken by each bowler
bowler_wickets = df_cleaned.groupby('Bowler')['Wkts'].sum().reset_index()

# Sort bowlers by the number of wickets taken
bowler_wickets_sorted = bowler_wickets.sort_values(by='Wkts', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(bowler_wickets_sorted['Bowler'], bowler_wickets_sorted['Wkts'], color='skyblue')
plt.title('Total Wickets Taken by Bowlers')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Generate Scatter Plot and Bar chart by using cleaned Batsman dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data loaded into a DataFrame named 'df'
# Read the bowlers dataset which is downloaded from github(File path)
df=pd.read_csv(r"C:\Users\navee\Downloads\Batsman_Data.csv\Batsman_Data_clean.csv")

print(df)
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd


# Remove rows with missing values in 'SR' and 'Runs' columns
df_cleaned = df.dropna(subset=['SR', 'Runs'])

# Convert 'SR' and 'Runs' columns to numeric (ignoring errors)
df_cleaned['SR'] = pd.to_numeric(df_cleaned['SR'], errors='coerce')
df_cleaned['Runs'] = pd.to_numeric(df_cleaned['Runs'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['SR'], df_cleaned['Runs'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Strike Rate vs Runs')
plt.xlabel('Strike Rate')
plt.ylabel('Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd

# Remove rows with missing values in 'Runs' and 'Batsman' columns
df_cleaned = df.dropna(subset=['Runs', 'Batsman'])

# Group by 'Batsman' and sum the total runs scored by each batsman
batsman_runs = df_cleaned.groupby('Batsman')['Runs'].sum().reset_index()

# Sort batsmen by the total runs scored
batsman_runs_sorted = batsman_runs.sort_values(by='Runs', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(batsman_runs_sorted['Batsman'], batsman_runs_sorted['Runs'], color='skyblue')
plt.title('Total Runs Scored by Batsmen')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
