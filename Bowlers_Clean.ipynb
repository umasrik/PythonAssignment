# Drop average Average and SR

```python
df.drop(['Ave', 'SR'], axis=1, inplace=True)
print(df)
```

# Apply clean logc on Bowler dataset

```python
df["Overs"] = df["Overs"].str.strip("-")
df["Mdns"] = df["Mdns"].str.strip("-")
df["Runs"] = df["Runs"].str.strip("-")
df["Wkts"] = df["Wkts"].str.strip("-")
df["Econ"] = df["Econ"].str.strip("-")
df
```

```python
df["Overs"] = df["Overs"].str.strip()
df["Mdns"] = df["Mdns"].str.strip()
df["Runs"] = df["Runs"].str.strip()
df["Wkts"] = df["Wkts"].str.strip()
df["Econ"] = df["Econ"].str.strip()
df
```

```python
df2=df["Overs"]==''
df2=df[~df2]
print(df2)
```
