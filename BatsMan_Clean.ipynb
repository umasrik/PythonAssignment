# Apply cleaning logic on Bats man dataset

```python
col1 = ["Bat1"]
df1[col1] = df1[col1].apply(lambda x:x.str.strip("absent"))
print(df1)
```

```python
col1 = ["Bat1"]
df1[col1] = df1[col1].apply(lambda x:x.str.strip("DNB"))
print(df1)
```

```python
col1 = ["Bat1"]
df1[col1] = df1[col1].apply(lambda x:x.str.strip("sub"))
print(df1)
```

```python
col1 = ["Bat1"]
df1[col1] = df1[col1].apply(lambda x:x.str.strip("TDNB"))
print(df1)
```

```python
col1 = ["Runs","BF","SR","4s","6s"]
df1[col1] = df1[col1].apply(lambda x:x.str.strip("-"))
print(df1)
```

```python
df3 = df1["Bat1"]==''
df3 = df1[~df3]
print(df3)
```

# Final cleaned Batsman dataset
